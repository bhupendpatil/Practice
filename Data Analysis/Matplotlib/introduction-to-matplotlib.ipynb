## Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = np.array([1,2,3,4])
y = np.array([11,22,33,44])

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="Simple plot",xlabel="x-axis",ylabel="y-axis")

# save and show
fig.savefig("same-plot.png")
```
