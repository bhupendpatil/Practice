## Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = np.array([1,2,3,4])
y = np.array([11,22,33,44])

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="Simple plot",xlabel="x-axis",ylabel="y-axis")

# save and show
fig.savefig("same-plot.png")
```

## Making figures with NumPy arrays

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot data
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# same data
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# make plot from dict
nut_butter_prices = {"Almond butter":10, "Peanut butter": 8,"Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Bhupend's Butter Store",ylabel="Price($)");
```
